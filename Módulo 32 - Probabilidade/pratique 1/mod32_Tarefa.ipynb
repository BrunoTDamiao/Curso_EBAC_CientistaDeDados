# Probabilidade - Tarefa

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import stats

```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
# Parâmetros da distribuição
mu = 15  # média
sigma = 0.1  # desvio padrão
x = 15.2  # valor limite

# Calculando a probabilidade de P(X > 15.2)
probability = 1 - stats.norm.cdf(x, loc=mu, scale=sigma)

print(f"A probabilidade de um pacote pesar mais de 15,2 kg é: {probability:.4f}")
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Parâmetros da distribuição
mu = 15  # média
sigma = 0.1  # desvio padrão
x = 0.95  # valor limite

# Cálculo do percentil de ordem 95%
percentile_95 = stats.norm.ppf(x, loc=mu, scale=sigma)

print(f"A probabilidade de um pacote pesar mais de 15,2 kg é: {percentile_95:.4f}")
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu código começa aqui
n = 5

# Calculando CDF para a Normal padrão
normal_cdf = stats.norm.cdf(Y)

# Calculando CDF para a t-Student
tstudent_cdf = stats.t.cdf(Y, df=n)

print("Probabilidades acumuladas para a Normal padrão (P(Y1 <= y)):")
print(normal_cdf)

print("\nProbabilidades acumuladas para a t-Student com 5 graus de liberdade (P(Y2 <= y)):")
print(tstudent_cdf)
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
# Plotando as probabilidades acumuladas
plt.plot(Y, normal_cdf, label="Normal padrão", color="blue", linewidth=2)
plt.plot(Y, tstudent_cdf, label="t-Student (df=5)", color="orange", linestyle="--", linewidth=2)

# Configurações do gráfico
plt.title("Distribuição Acumulada: Normal Padrão vs t-Student", fontsize=14)
plt.xlabel("Valores de Y", fontsize=12)
plt.ylabel("Probabilidade Acumulada", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.4)

# Exibindo o gráfico
plt.show()
```

5. 3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu código começa aqui
n = 5

# Calculando PPF para a Normal padrão
normal_cdf = stats.norm.ppf(Y)

# Calculando PPF para a t-Student
tstudent_cdf = stats.t.ppf(Y, df=n)

print("Probabilidades acumuladas para a Normal padrão (P(Y1 <= y)):")
print(normal_cdf)

print("\nProbabilidades acumuladas para a t-Student com 5 graus de liberdade (P(Y2 <= y)):")
print(tstudent_cdf)
```

6. Faça um gráfico com os dados do item anterior.

```python
# Plotando as probabilidades acumuladas
plt.plot(Y, normal_cdf, label="Normal padrão", color="blue", linewidth=2)
plt.plot(Y, tstudent_cdf, label="t-Student (df=5)", color="orange", linestyle="--", linewidth=2)

# Configurações do gráfico
plt.title("Distribuição Acumulada: Normal Padrão vs t-Student", fontsize=14)
plt.xlabel("Valores de Y", fontsize=12)
plt.ylabel("Probabilidade Acumulada", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.4)

# Exibindo o gráfico
plt.show()
```
