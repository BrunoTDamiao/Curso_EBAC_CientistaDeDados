# Probabilidade - Tarefa

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import stats

```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
# Parmetros da distribuio
mu = 15  # mdia
sigma = 0.1  # desvio padro
x = 15.2  # valor limite

# Calculando a probabilidade de P(X > 15.2)
probability = 1 - stats.norm.cdf(x, loc=mu, scale=sigma)

print(f"A probabilidade de um pacote pesar mais de 15,2 kg : {probability:.4f}")
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Parmetros da distribuio
mu = 15  # mdia
sigma = 0.1  # desvio padro
x = 0.95  # valor limite

# Clculo do percentil de ordem 95%
percentile_95 = stats.norm.ppf(x, loc=mu, scale=sigma)

print(f"A probabilidade de um pacote pesar mais de 15,2 kg : {percentile_95:.4f}")
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu cdigo comea aqui
n = 5

# Calculando CDF para a Normal padro
normal_cdf = stats.norm.cdf(Y)

# Calculando CDF para a t-Student
tstudent_cdf = stats.t.cdf(Y, df=n)

print("Probabilidades acumuladas para a Normal padro (P(Y1 <= y)):")
print(normal_cdf)

print("\nProbabilidades acumuladas para a t-Student com 5 graus de liberdade (P(Y2 <= y)):")
print(tstudent_cdf)
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
# Plotando as probabilidades acumuladas
plt.plot(Y, normal_cdf, label="Normal padro", color="blue", linewidth=2)
plt.plot(Y, tstudent_cdf, label="t-Student (df=5)", color="orange", linestyle="--", linewidth=2)

# Configuraes do grfico
plt.title("Distribuio Acumulada: Normal Padro vs t-Student", fontsize=14)
plt.xlabel("Valores de Y", fontsize=12)
plt.ylabel("Probabilidade Acumulada", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.4)

# Exibindo o grfico
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)
# Seu cdigo comea aqui
n = 5

# Calculando PPF para a Normal padro
normal_cdf = stats.norm.ppf(Y)

# Calculando PPF para a t-Student
tstudent_cdf = stats.t.ppf(Y, df=n)

print("Probabilidades acumuladas para a Normal padro (P(Y1 <= y)):")
print(normal_cdf)

print("\nProbabilidades acumuladas para a t-Student com 5 graus de liberdade (P(Y2 <= y)):")
print(tstudent_cdf)
```

6. Faa um grfico com os dados do item anterior.

```python
# Plotando as probabilidades acumuladas
plt.plot(Y, normal_cdf, label="Normal padro", color="blue", linewidth=2)
plt.plot(Y, tstudent_cdf, label="t-Student (df=5)", color="orange", linestyle="--", linewidth=2)

# Configuraes do grfico
plt.title("Distribuio Acumulada: Normal Padro vs t-Student", fontsize=14)
plt.xlabel("Valores de Y", fontsize=12)
plt.ylabel("Probabilidade Acumulada", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.4)

# Exibindo o grfico
plt.show()
```
