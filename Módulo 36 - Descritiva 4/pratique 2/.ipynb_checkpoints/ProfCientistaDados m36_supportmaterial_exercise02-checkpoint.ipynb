## Criando variveis

**Variveis temporais**:  possvel que haja alguma regra de liberao de crditos de acordo com o dia do ms, o dia da semana, fim de semana (sim ou no), ou at alteraes temporais da regra. Vamos criar as seguintes variveis para testar estas hipteses:

- Dia da semana
- Dummy para fim de semana
- Dia do ms
- Ms (j na base)
- Trimestre
- Ano (j na base)

**Valor da nota**: Vamos criar tambm algumas variveis para resumir o valor da nota, e posteriormente criar uma intuio sobre categorizaes de variveis contnuas.

- Crie categorizaes da varivel ```Valor NF```, por quants, em 5, 10, 20 e 50 categorias.

````python
df_raw['resposta'] = np.where(df_raw['flag_credito'] > 0,1,0)
````

````python
# Converso para datetime e filtragem por Data 
df_raw['Data Emisso'] = pd.to_datetime(df_raw['Data Emisso'])
df = df_raw[df_raw['Data Emisso'] >= '2020-01-01'].copy()

# Extrao de informaes temporais
df['dia_semana'] = df['Data Emisso'].dt.dayofweek  # 0 = Segunda, ..., 6 = Domingo
df['fim_semana'] = df['dia_semana'].isin([5, 6]).astype(int)  # Fim de semana = Sbado/Domingo
df['dia_mes'] = df['Data Emisso'].dt.day
df['trimestre'] = df['Data Emisso'].dt.quarter
df['mes'] = df['Data Emisso'].dt.month
df['ano'] = df['Data Emisso'].dt.year

quantis = [5, 10, 20, 50]
for q in quantis:
    df[f'valor_nf_quantis_{q}'] = pd.qcut(df['Valor NF'], q, labels=False)

df.head()
````

## Calcule o IV de todas essas variveis

Sugesto: Construa um metadados e tente utilizar uma funo como a que foi utilizada durante a aula. Voc vai conseguir fazer esse clculo para as variveis uma a uma, mas em casos reais em que vamos mais alm na criao de variveis, podemos ter centenas, ou at milhares de variveis, esses clculos *precisam* ser automticos nesses casos.

````python
variaveis= [
    'ano', 'mes', 'dia_semana', 'fim_semana', 'dia_mes', 'trimestre', 'valor_nf_quantis_5',
    'valor_nf_quantis_10', 'valor_nf_quantis_20', 'valor_nf_quantis_50'
]
variaveis

metadados = pd.DataFrame({
    'variavel': variaveis,
    'tipo': [df[var].dtype for var in variaveis]
})

metadados['papel'] = 'covariavel'

metadados['nunique'] = [df[var].nunique() for var in variaveis]
````

````python
# Funo para calcular o IV
def IV(variavel, resposta):
    tab = pd.crosstab(variavel, resposta, margins=True, margins_name='total')

    rotulo_evento = tab.columns[0]
    rotulo_nao_evento = tab.columns[1]

    tab['pct_evento'] = tab[rotulo_evento] / tab.loc['total', rotulo_evento]
    tab['pct_nao_evento'] = tab[rotulo_nao_evento] / tab.loc['total', rotulo_nao_evento]
    tab['woe'] = np.log(tab['pct_evento'] / tab['pct_nao_evento'])
    tab['iv_parcial'] = (tab['pct_evento'] - tab['pct_nao_evento']) * tab['woe']
    return tab['iv_parcial'].sum()
````

````python
# Inicializa um dicionrio para armazenar os IVs
iv_results = {}
resposta = df['resposta']

for _, row in metadados.iterrows():
    var = row['variavel']
    try:
        nunique = row['nunique']
        if nunique > 5:
            iv_results[var] = IV(pd.qcut(df[var], 5, duplicates='drop'), resposta)
        else:
            iv_results[var] = IV(df[var], resposta)
    except Exception as e:
        print(f"Erro ao calcular IV para a varivel '{var}': {e}")
        iv_results[var] = None  # Marca erro com valor None

# Atualiza o dataframe de uma vez
metadados['IV'] = metadados['variavel'].map(iv_results)

# Retorna o dataframe atualizado
metadados
````

## Avalie o WOE e o IV no tempo

Avalie o WOE e o IV no tempo da varivel ```categoria```.

````python
def calc_woe_iv(df, var, target):
   
    total_eventos = df[target].sum()
    total_nao_eventos = df[target].count() - total_eventos    
    grupo = df.groupby(var)[target].agg(['sum', 'count']).reset_index()
    grupo.columns = [var, 'eventos', 'total']
    grupo['nao_eventos'] = grupo['total'] - grupo['eventos']
    grupo['perc_eventos'] = grupo['eventos'] / total_eventos
    grupo['perc_nao_eventos'] = grupo['nao_eventos'] / total_nao_eventos
    
    # WOE
    grupo['woe'] = np.log(grupo['perc_eventos'] / grupo['perc_nao_eventos'])    
    # IV parcial
    grupo['iv_parcial'] = (grupo['perc_eventos'] - grupo['perc_nao_eventos']) * grupo['woe']    
    # IV total
    iv = grupo['iv_parcial'].sum()
    
    return grupo[[var, 'woe', 'iv_parcial']], iv

````

````python
# Lista para armazenar os dados
woe_iv_tempo = []

# Iterando sobre os trimestres
for trimestre in df['trimestre'].unique():
    df_trim = df[df['trimestre'] == trimestre]
    woe_iv, iv = calc_woe_iv(df_trim, 'categoria', 'resposta')
    
    # Adicionando os dados de WOE e IV para cada categoria
    woe_iv_tempo.extend([{'trimestre': trimestre, 'categoria': row['categoria'], 'woe': row['woe'], 'iv': iv}
                         for _, row in woe_iv.iterrows()])

# Criando o DataFrame com os dados
woe_iv_df = pd.DataFrame(woe_iv_tempo)
````

````python
plt.figure(figsize=(12, 6))
sns.lineplot(data=woe_iv_df, x='trimestre', y='woe', hue='categoria', marker='o')

# Configuraes adicionais
plt.xlabel('Trimestre')
plt.ylabel('Weight of Evidence (WOE)')
plt.title('WOE da Varivel Categoria por Trimestre')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Categoria', bbox_to_anchor=(1.05, 1), loc='upper left')

# Exibe o grfico
plt.show()
````

````python
plt.figure(figsize=(8, 4))
sns.lineplot(data=woe_iv_df, x='trimestre', y='iv', marker='o')

# Definindo configuraes do grfico
plt.title('IV da Varivel Categoria por Trimestre')
plt.xlabel('Trimestre')
plt.ylabel('Information Value (IV)')
plt.ylim(0, 0.6)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Exibe o grfico
plt.show()
````
