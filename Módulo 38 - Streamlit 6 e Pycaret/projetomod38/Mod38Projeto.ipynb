## Desenvolvimento do modelo

Desenvolva um modelo de *credit scoring* atravs de uma regresso logstica.

- Trate valores missings e outliers
- Trate 'zeros estruturais'
- Faa agrupamentos de categorias conforme vimos em aula
- Proponha uma equao preditiva para 'mau'
- Caso hajam categorias no significantes, justifique

````python
df.isnull().sum()
````

````python
# Excluir a varivel 'posse_de_veiculo' (no discrimina)
df.drop(columns=['posse_de_veiculo'], inplace=True)

# Agrupar quantidade de filhos acima de 5 em uma nica categoria "6+"
df['qtd_filhos'] = df['qtd_filhos'].apply(lambda x: '<6' if x <= 5 else '6+')

# Remover categorias 'Pensionista' e 'Bolsista' da varivel 'tipo_renda'
df = df[~df['tipo_renda'].isin(['Pensionista', 'Bolsista'])]

# Agrupar "Ps-graduao" em "Superior ou mais" na varivel 'educacao'
df['educacao'] = df['educacao'].replace({'Ps-graduao': 'Superior ou mais'})

# Criar nova varivel 'faixa_idade', categorizando como "<30" e "30+"
df['faixa_idade'] = df['idade'].apply(lambda x: '<30' if x < 30 else '30+')

# Agrupar tempo de emprego acima de 25 anos em "25+ anos"
df['tempo_emprego'] = df['tempo_emprego'].apply(lambda x: "<25" if x < 25 else '25+ anos')

# Agrupar quantidade de pessoas na residncia acima de 6 em "7+"
df['qt_pessoas_residencia'] = df['qt_pessoas_residencia'].apply(lambda x: '<7' if x <= 6 else '7+')

# Limitar a renda a valores de at 800.000 para tratar outliers
df = df[df['renda'] <= 800000]
````

````python
df.head()
````

````python
# Ordenando os dados por data de referncia
df_sorted = df.sort_values(by='data_ref')

# Definindo o perodo de corte para os ltimos 3 meses
cutoff_date = df_sorted['data_ref'].max() - pd.DateOffset(months=3)

# Dividindo os dados em treino e teste com base na data
train_df = df_sorted[df_sorted['data_ref'] <= cutoff_date]
test_df = df_sorted[df_sorted['data_ref'] > cutoff_date]

# Criando variveis dummy para as variveis categricas, excluindo as colunas 'mau', 'data_ref' e 'index'
X_train = pd.get_dummies(train_df.drop(columns=['mau', 'data_ref', 'index']))
X_test = pd.get_dummies(test_df.drop(columns=['mau', 'data_ref', 'index']))

# Garantindo que X_train e X_test tenham as mesmas colunas
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)

# Definindo a varivel resposta para treino e teste
y_train = train_df['mau']
y_test = test_df['mau']
````

````python
# Ajustando o modelo de Regresso Logstica com class_weight='balanced'
model = LogisticRegression(max_iter=3000, class_weight='balanced', random_state=66)
model.fit(X_train, y_train)
````

````python
# Ajustando o limiar para a classificao
threshold = 0.69  # Valor ajustado

# Avaliao na base de treino
y_prob_train = model.predict_proba(X_train)[:, 1]  # Probabilidades para a classe positiva na base de treino
y_pred_train_adjusted = (y_prob_train > threshold).astype(int)  # Aplicando o limiar ajustado

# Avaliao na base de teste
y_prob_test = model.predict_proba(X_test)[:, 1]  # Probabilidades para a classe positiva na base de teste
y_pred_test_adjusted = (y_prob_test > threshold).astype(int)  # Aplicando o limiar ajustado

# Matriz de Confuso e Relatrio de Classificao para a base de treino
conf_matrix_train = confusion_matrix(y_train, y_pred_train_adjusted)
class_report_train = classification_report(y_train, y_pred_train_adjusted, target_names=['Classe 0', 'Classe 1'])

# Matriz de Confuso e Relatrio de Classificao para a base de teste
conf_matrix_test = confusion_matrix(y_test, y_pred_test_adjusted)
class_report_test = classification_report(y_test, y_pred_test_adjusted, target_names=['Classe 0', 'Classe 1'])

# Exibindo resultados de forma mais legvel
print("Avaliao na Base de Treino:")
print("Matriz de Confuso:")
print(conf_matrix_train)
print("\nRelatrio de Classificao:")
print(class_report_train)

print("\nAvaliao na Base de Teste:")
print("Matriz de Confuso:")
print(conf_matrix_test)
print("\nRelatrio de Classificao:")
print(class_report_test)
````

## Avaliao do modelo

Avalie o poder discriminante do modelo pelo menos avaliando acurcia, KS e Gini.

Avalie estas mtricas nas bases de desenvolvimento e *out of time*.

````python
def ks_statistic(y_true, y_prob):
    """Calcula a estatstica KS"""
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    return max(tpr - fpr)

def gini_coefficient(y_true, y_prob):
    """Calcula o coeficiente Gini"""
    return 2 * roc_auc_score(y_true, y_prob) - 1

# Mtricas na base de treino
accuracy_train = accuracy_score(y_train, y_pred_train_adjusted)
ks_train = ks_statistic(y_train, y_prob_train)
gini_train = gini_coefficient(y_train, y_prob_train)

# Mtricas na base de teste (OOT)
accuracy_test = accuracy_score(y_test, y_pred_test_adjusted)
ks_test = ks_statistic(y_test, y_prob_test)
gini_test = gini_coefficient(y_test, y_prob_test)

# Exibindo os resultados de forma mais legvel e formatada
print("Avaliao na Base de Treino:")
print(f"Acurcia: {accuracy_train:.4f}")
print(f"KS: {ks_train:.4f}")
print(f"Gini: {gini_train:.4f}")

print("\nAvaliao na Base Out-of-Time (Teste):")
print(f"Acurcia: {accuracy_test:.4f}")
print(f"KS: {ks_test:.4f}")
print(f"Gini: {gini_test:.4f}")
````

### Substituio de nulos (nans)

Existe nulos na base?  dado numrico ou categrico? qual o valor de substituio? mdia? valor mais frequente? etc

> Os dados nulos so relacionados a vriavel "tempo_emprego" na categoria "pensionista". Irei eliminar os dados nulos pois faz mais sentido.


````python
# 1. Substituio de nulos
def substituir_nulos(df):
    df.dropna(inplace=True)
    return df
````

### Remoo de outliers

Como identificar outlier? Substituir o outlier por algum valor? Remover a linha?

> Os outliers so os dados que esto fora do intervalo entre 1,5 vezes o IQR abaixo do primeiro quartil (Q1) e 1,5 vezes o IQR acima do terceiro quartil (Q3).

````python
def remover_outliers(df):
    df = df.copy()
    for col in df.select_dtypes(include=[np.number]).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        df = df[(df[col] >= (Q1 - 1.5 * IQR)) & (df[col] <= (Q3 + 1.5 * IQR))]
    return df
````

### Seleo de variveis

Qual tipo de tcnica? Boruta? Feature importance? 

````python
# 3. Seleo de variveis usando feature importance
def selecionar_variaveis(X, y, n_features=8):
    model = RandomForestClassifier()
    model.fit(X, y)
    importances = model.feature_importances_
    indices = np.argsort(importances)[-n_features:]  # Seleciona os ndices das n_features mais importantes
    selected_columns = X.columns[indices]  # Nomes das variveis selecionadas
    return X[selected_columns], selected_columns
````

### Reduo de dimensionalidade (PCA)

Aplicar PCA para reduzir a dimensionalidade para 5

````python
# 4. Aplicao de PCA
def aplicar_pca(X, n_components=5):
    pca = PCA(n_components=n_components)
    X_pca = pca.fit_transform(X)
    return pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(n_components)])
````

Aplicar o get_dummies() ou onehotencoder() para transformar colunas catgoricas do dataframe em colunas de 0 e 1.

````python
# 5. Criao de dummies
def criar_dummies(df):
    return pd.get_dummies(df, drop_first=True)
````

### Pipeline 

Crie um pipeline contendo essas funes.

preprocessamento()
- substituicao de nulos
- remoo outliers
- PCA
- Criao de dummy de pelo menos 1 varivel (posse_de_veiculo)

````python
def preprocessamento(df, target_col='mau', drop_cols=None, n_features=8, n_components=5):
    """
    Pipeline de pr-processamento de dados.
    
    Args:
        df (pd.DataFrame): DataFrame de entrada.
        target_col (str): Nome da varivel alvo. Padro: 'mau'.
        drop_cols (list): Colunas a serem removidas antes do preprocessamento. Padro: ['index', 'data_ref'].
        n_features (int): Nmero de variveis mais importantes a selecionar. Padro: 8.
        n_components (int): Nmero de componentes principais a manter no PCA. Padro: 5.
    
    Returns:
        pd.DataFrame: DataFrame preprocessado com PCA aplicado.
        list: Lista das colunas selecionadas aps a seleo de variveis.
    """
    if drop_cols is None:
        drop_cols = ['index', 'data_ref']
    
    # Separar varivel alvo e remover colunas indesejadas
    y = df[target_col].copy()
    X = df.drop(columns=[target_col] + drop_cols, errors='ignore')
    
    # Substituio de Nulos
    print("Etapa 1: Substituio de Nulos")
    X = substituir_nulos(X)
    y = y[X.index]  # Manter apenas ndices vlidos
    
    # Remoo de Outliers
    print("Etapa 2: Remoo de Outliers")
    indices_antes = X.index
    X = remover_outliers(X)
    y = y[X.index]  # Atualizar y com base nos ndices restantes em X
    
    # Criao de Dummies
    print("Etapa 3: Criao de Dummies")
    X = criar_dummies(X)
    
    # Verificao de consistncia aps pr-processamento
    if X.shape[0] != y.shape[0]:
        raise ValueError("Nmero de amostras em X e y no coincide aps pr-processamento")

    # Seleo de Variveis
    print("Etapa 4: Seleo de Variveis")
    X_selecionado, colunas_selecionadas = selecionar_variaveis(X, y, n_features=n_features)
    
    # Verificao de consistncia aps seleo de variveis
    if X_selecionado.shape[0] != y.shape[0]:
        raise ValueError("Nmero de amostras em X e y no coincide aps seleo de variveis")

    # Aplicao de PCA
    print("Etapa 5: Aplicao de PCA")
    X_pca = aplicar_pca(X_selecionado, n_components=n_components)

    # Adicionar varivel alvo de volta
    X_pca[target_col] = y.reset_index(drop=True)
    
    return X_pca, colunas_selecionadas
````

````python
df_processado, variaveis_selecionadas = preprocessamento(df)
print(df_processado.head())
print("Variveis selecionadas:", variaveis_selecionadas)
````
